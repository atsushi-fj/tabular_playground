```python
import lightgbm as lgb
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#データセットの読み込み
train_dataset = pd.read_csv('train.csv')
```

```python
train_dataset
```

```python
#独立変数と従属変数に分ける
X1 = train_dataset.iloc[:, 1:-2].values
y1 = train_dataset.iloc[:, -2].values
```

```python
X1
```

```python
y1
```

```python
X1[0, :]
```

```python
#フィチャースケーリング
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X1[:, -3:-2] = sc.fit_transform(X1[:, -3:-2])
```

```python
X1[0, :]
```

```python
#データセットの分割
from sklearn.model_selection import train_test_split
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=0)
```

```python
#モデルの訓練
gbm = lgb.LGBMClassifier()
gbm.fit(X1_train, y1_train)
```

```python
#結果の予測
y1_pred = gbm.predict(X1_test)
y1_pred_prob = gbm.predict_proba(X1_test)
```

```python
df_pred = pd.DataFrame({'target':y1_test, 'target_pred':y1_pred})
display(df_pred)
```

```python
df_pred_prob = pd.DataFrame({'target':y1_test, 'target0_prob':y1_pred_prob[:, 0], 'target1_prob':y1_pred_prob[:, 1]})
display(df_pred_prob)
```

```python
#モデルの評価
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y1_test, y1_pred)
print(cm)
accuracy_score(y1_test, y1_pred)
```
